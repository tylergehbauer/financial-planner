## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
btc_request = requests.get(btc_url).json() #print(btc_request) shows the dict of dicts
btc_price = btc_request['data']['1']['quotes']['USD']['price']
print(btc_price) #to check if I located the price properly

# Fetch current ETH price
eth_request = requests.get(eth_url).json()
eth_price = eth_request['data']['1027']['quotes']['USD']['price']
print(eth_price)
                                                
# Compute current value of my crpto
my_btc_value = my_btc * btc_price
my_eth_value = my_eth * eth_price

# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")


# Create the Alpaca API object
api = tradeapi.REST( #object im going to use when using alpaca sdk
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
#print(alpaca_api_key)
```

```python
# Format current date as ISO format
current_date = pd.Timestamp("2021-12-31", tz="America/New_York").isoformat()
#2022-month-day

# Set the tickers
tickers = ["AGG", "SPY"] #spy is a stock and agg is abond

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
df_portfolio = api.get_barset(      
    tickers,
    timeframe, 
    start = current_date,
    end = current_date,
    limit = 1000).df 

# Preview DataFrame
df_portfolio
```

```python
# Pick AGG and SPY close prices
agg_close_price = float(df_portfolio['AGG']['close'])
spy_close_price = float(df_portfolio['SPY']['close']) #use float to get JUST the close value 

# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
my_agg_value = my_agg * agg_close_price
my_spy_value = my_spy * spy_close_price

# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
monthly_income = 12000

# Consolidate financial assets data
total_shares = my_spy_value + my_agg_value
total_crypto = my_eth_value + my_btc_value
value_data = { 'amount': [total_crypto, total_shares]} #these are what are in the amount column

# Create savings DataFrame
df_savings = pd.DataFrame(value_data, index = ['crypto','shares']) #this makes the index 'crypto' and 'shares' 
#while initializing the new dataframe

# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
df_savings.plot.pie( y = 'amount', title='Composition of Personal Savings')
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3
#ideal should be 3 times what you make every month

# Calculate total amount of savings
total_savings = df_savings['amount'][0] + df_savings['amount'][1]

#total_savings= 6000 #to check my variables works 

else_variable = emergency_fund - total_savings

# Validate saving health
if total_savings > emergency_fund:
    print('Congratulations! You have enough money in your emergency fund.')
elif total_savings == emergency_fund:
    print('Congratulations! You have met your financial goal.')
elif total_savings < emergency_fund:
    print(f'You need ${else_variable} more to reach your financial goal.')
else:
    print('Something is just not right here.')
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2016-12-31', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-12-31', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
df_stock_data = api.get_barset(      
    tickers, #same tickers as before
    timeframe, #same timeframe as before
    start = start_date, #only thing we are really changing are the dates
    end = end_date,
    limit = 1000).df 


# Display sample data
df_stock_data.head() #1000/252 (252 trading days) is 4 years worth of data
#i.e 2018,2019,2020,2021
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns

MC_stock_data = MCSimulation( #MCSimulation is the function
    portfolio_data = df_stock_data, #uses the dataframe we just created above
    weights = [.40, .60], #assignment wants 60% SPY and 40% AGG, so weights are based on the columns above (alpabetical order)
    num_simulation = 500, #directions (number of simulations)
    num_trading_days = 252*30 #30 years of trading
)
```

```python
# Printing the simulation input data
MC_stock_data.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_stock_data.calc_cumulative_return()
#.calc_cumulative_return() runs the monte carlo simulation using the MCforecast.py
```

```python
# Plot simulation outcomes
line_plot = MC_stock_data.plot_simulation()
#think of this as a 'multiverse' of possibilites of returns over the next 7560 days
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_stock_data.plot_distribution()
#.plot_distribution is the command that adds the 95% confidence interval
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
summary = MC_stock_data.summarize_cumulative_return() 
#.summarize_cumulative_returns() is the function that gives us the information below
# Print summary statistics
print(summary)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(summary[8] * initial_investment, 2) #This locates the upper and lower confidence interval 
ci_upper = round(summary[9] * initial_investment, 2) #and multiplies it by our initial investment and rounds the answer

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
larger_initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
larger_ci_lower = round(summary[8] * larger_initial_investment, 2) #same exact formula as before but with different varibale names 
larger_ci_upper = round(summary[9] * larger_initial_investment, 2)

# Print results
print(f"There is a 95% chance that an initial investment of ${larger_initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${larger_ci_lower} and ${larger_ci_upper}")
```

### Five Years Retirement Option

```python
# Configuring a Monte Carlo simulation to forecast 5 years cumulative returns
#We will have a higer stock to bond ratio in the weights (means more risk), with a larger initial investment
MC_five_year = MCSimulation( #MCSimulation is the function
    portfolio_data = df_stock_data, #uses the dataframe we created in part 2
    weights = [.30, .70], #we will adjust the weight to be 70% SPY and 30% AGG, so more weight for the stock SKY
    num_simulation = 500, #directions (number of simulations)
    num_trading_days = 252*5 #5 years
)
```

```python
# Running a Monte Carlo simulation to forecast 5 years cumulative returns
MC_five_year.calc_cumulative_return()
```

```python
# Plot simulation outcomes
five_year_line_plot = MC_five_year.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
five_year_dist_plot = MC_five_year.plot_distribution()
```

```python
# Fetch summary statistics from the Monte Carlo simulation results
five_year_summary = MC_five_year.summarize_cumulative_return()

# Print summary statistics
print(five_year_summary)
```

```python
# Set initial investment
five_year_initial_investment = 60000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $60,000
ci_lower_five = round(five_year_summary[8]*60000,2) #8th index entry is CI lower (starting at 0)
ci_upper_five = round(five_year_summary[9]*60000,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${five_year_initial_investment} in the portfolio"
      f" over the next 5 years will end within in the range of"
      f" ${ci_lower_five} and ${ci_upper_five}")
```

This seems like a decent option for a short term retirement plan, but does involve the risk of losing money. Maybe would be better if I adjusted the weights differently.
